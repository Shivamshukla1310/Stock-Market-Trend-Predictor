```python
# Training And Testing

data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])
```

```python
data_training.shape
```

```python
data_testing.shape
```

```python
from sklearn .preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
```

```python
data_training_array.shape
```

```python
data_training_array
```

```python
x_train = []
y_train = []

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
# Model Building 
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

LSTM Input --> 3D Array it consist of (Batch_size, time_steps, seq_len) --> Sequence Length(seq_len)
Here Batch Size is nothing but how many batches of file is coming here 
time steps just tell the steps of time t1, t2, t3 etc it is basically used in LSTM for time series Forecasting
Output Can Be 3D Or 2D It depends on returning sequence
2D -> (batch_sizem units)
3D -> (batch_size, time_steps, units)

```python
model = Sequential()

model.add(LSTM(units = 50, activation = "relu", return_sequences = True, input_shape = (x_train.shape[1],1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 60, activation = "relu", return_sequences = True))
model.add(Dropout(0.3))

model.add(LSTM(units = 80, activation = "relu", return_sequences = True))
model.add(Dropout(0.4))

model.add(LSTM(units = 120, activation = "relu"))
model.add(Dropout(0.5))

model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
model.compile(optimizer = "adam", loss = "mean_squared_error")
model.fit(x_train, y_train, epochs = 50)
```

```python
past_100_days = data_training.tail(100)
```

```python
# final_df = past_100_days.append(data_testing, ignore_index = True) --> in recent update after v2.0 DataFrame.append was deprecated From it
final_df = pd.concat([past_100_days, data_testing], ignore_index=True) # We Can use this instead both does the same job
```

```python
final_df.head()
```

```python
input_data = scaler.fit_transform(final_df)
```

```python
x_test = []
y_test = []

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i, 0])

x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
x_test.shape
```

```python
y_predicted  = model.predict(x_test)
```

```python
y_predicted.shape
```

```python
scaler.scale_
```

```python
scaler_factor = 1 / 0.00345503
y_predicted = y_predicted * scaler_factor
y_test = y_test * scaler_factor
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test, label = "Original Price", linewidth = 0.5)  
plt.plot(y_predicted, label = "Predicted Price", linewidth = 1)
plt.legend()
plt.show()
```
